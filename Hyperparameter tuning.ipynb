```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
# number of rows and Columns in this dataset
data_frame.shape
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# checking the distribution of Target Varibale
data_frame['label'].value_counts()
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
X
```

```python
Y
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

```python
#GRIDSEARCHCV
```

```python
# loading the Support Vector Machine model
model = SVC()
```

```python
#We have hyper parameter tuning using GRIDSEARCH, BAYESIAN NETWORK, RaNDOM 
param_grid = {'kernel':['linear','poly','rbf','sigmoid'],
              'C':[1, 5, 10, 20],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
             }
# /*   higher value of C will lead to a more accurate model, but it may also lead to overfitting. A lower value of C will lead to a less accurate model, but it may also be less likely to overfit.

# The value of gamma controls the smoothness of the decision boundary. 

# The choice of kernel can also affect the accuracy and performance of the model. The RBF kernel is a popular choice for SVMs, as it is generally effective for a wide range of problems. However, other kernels may be more appropriate for certain types of data.  */
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3, cv=5)
  
# fitting the model for grid search
grid.fit(X, Y)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
